```python
import requests
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# URL for the COVID-19 data in India (Data till August 2021)
url = "https://api.covid19india.org/csv/latest/state_wise.csv"
```

```python
# Send a GET request to the URL
response = requests.get(url)
```

```python
#Read the response as a text
data = response.text
```

```python
# Write the response to a CSV file
with open("covid_data.csv", "w") as file:
    file.write(data)
```

```python
# Read the CSV file using pandas
df = pd.read_csv("covid_data.csv")
df
```

```python
# Filter necessary columns
df = df[df['State'] != 'State Unassigned']
df = df[df['State'] != 'Total']
df = df[["State", "Confirmed", "Deaths", "Recovered", "Active"]]

# Print the data
df
```

```python
# Basic data exploration
print(df.head())
print('\n')
print(df.describe())
```

```python
#Data exploration
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Confirmed', data=df)
plt.xticks(rotation=90)
plt.title('Confirmed COVID-19 Cases by State')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Deaths', data=df)
plt.xticks(rotation=90)
plt.title('COVID-19 Deaths by State')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Recovered', data=df)
plt.xticks(rotation=90)
plt.title('COVID-19 Recovered Cases by State')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Active', data=df)
plt.xticks(rotation=90)
plt.title('Active COVID-19 Cases by State')
plt.show()
```

```python
# Calculate Mortality Rate and Recovery Rate
df['Mortality Rate'] = (df['Deaths'] / df['Confirmed']) * 100
df['Recovery Rate'] = (df['Recovered'] / df['Confirmed']) * 100

# Sort the DataFrame by Confirmed cases in descending order
df_sorted_confirmed = df.sort_values(by='Confirmed', ascending=False)

# Get the top 5 and bottom 5 states by Confirmed cases
top_5_states_confirmed = df_sorted_confirmed.head(5)
bottom_5_states_confirmed = df_sorted_confirmed.tail(5)

# Visualize Mortality Rate and Recovery Rate
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Mortality Rate', data=df)
plt.xticks(rotation=90)
plt.title('COVID-19 Mortality Rate by State')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Recovery Rate', data=df)
plt.xticks(rotation=90)
plt.title('COVID-19 Recovery Rate by State')
plt.show()

# Display the top 5 and bottom 5 states by Confirmed cases
print("Top 5 States by Confirmed Cases:")
print(top_5_states_confirmed[['State', 'Confirmed']])

print("\nBottom 5 States by Confirmed Cases:")
print(bottom_5_states_confirmed[['State', 'Confirmed']])
```

```python
#Top 10 states with highest confirmed cases
top_10_confirmed = df.nlargest(10, 'Confirmed')

plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Confirmed', data=top_10_confirmed)
plt.xticks(rotation=45)
plt.title('Top 10 States with Highest Confirmed Cases')
plt.show()
```

```python
# Calculate and visualize the correlation between columns
correlation_matrix = df_sorted_confirmed.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Create a hypothetical "Population Density" column
# Population density = (Confirmed / Population) * 1000 (for simplicity)
# We'll use a constant for population for this example.
population = 1000000  # Constant population value for illustration
df_sorted_confirmed['Population Density'] = (df_sorted_confirmed['Confirmed'] / population) * 1000

# Scatter plot of Active Cases vs. Population Density
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Population Density', y='Active', data=df_sorted_confirmed)
plt.title('Active Cases vs. Population Density')
plt.xlabel('Population Density (per 1000)')
plt.ylabel('Active Cases')
plt.show()

# Scatter plot of Mortality Rate vs. Recovery Rate
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Mortality Rate', y='Recovery Rate', data=df_sorted_confirmed)
plt.title('Mortality Rate vs. Recovery Rate')
plt.xlabel('Mortality Rate (%)')
plt.ylabel('Recovery Rate (%)')
plt.show()
```

```python
# Identify the state with the highest and lowest mortality rates
state_highest_mortality = df_sorted_confirmed[df_sorted_confirmed['Mortality Rate'] == df_sorted_confirmed['Mortality Rate'].max()]['State'].values[0]
state_lowest_mortality = df_sorted_confirmed[df_sorted_confirmed['Mortality Rate'] == df_sorted_confirmed['Mortality Rate'].min()]['State'].values[0]

# Calculate average and median values
average_values = df_sorted_confirmed[['Confirmed', 'Deaths', 'Recovered', 'Active']].mean()
median_values = df_sorted_confirmed[['Confirmed', 'Deaths', 'Recovered', 'Active']].median()

# Calculate variance and standard deviation
variance_values = df_sorted_confirmed[['Confirmed', 'Deaths', 'Recovered', 'Active']].var()
std_deviation_values = df_sorted_confirmed[['Confirmed', 'Deaths', 'Recovered', 'Active']].std()

# Display the results
print(f"State with the highest mortality rate: {state_highest_mortality}")
print(f"State with the lowest mortality rate: {state_lowest_mortality}")

print("\nAverage Values:")
print(average_values)

print("\nMedian Values:")
print(median_values)

print("\nVariance Values:")
print(variance_values)

print("\nStandard Deviation Values:")
print(std_deviation_values)
```

```python
import scipy.stats as stats
import numpy as np

# Select the column for which you want to create a normal distribution
data_column = df_sorted_confirmed['Confirmed']

# Calculate mean and standard deviation of the data
mean = data_column.mean()
std_dev = data_column.std()

# Generate values for the x-axis (range of confirmed cases)
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 1000)

# Calculate the PDF (Probability Density Function) of the normal distribution
pdf = stats.norm.pdf(x, mean, std_dev)

# Plot the normal distribution curve
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Normal Distribution', color='blue')
plt.xlabel('Confirmed Cases')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Confirmed COVID-19 Cases')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# Load the COVID-19 data

# Select relevant columns for clustering (e.g., Confirmed, Deaths, Recovered, Active)
data_for_clustering = df_sorted_confirmed[['Confirmed', 'Deaths', 'Recovered', 'Active']]

# Perform hierarchical clustering using Ward's linkage method
linkage_matrix = linkage(data_for_clustering, method='ward', metric='euclidean')

# Create a dendrogram to visualize the hierarchy of clusters
plt.figure(figsize=(12, 8))
dendrogram(linkage_matrix, orientation="top", labels=df_sorted_confirmed['State'].tolist(), distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('States')
plt.ylabel('Distance')
plt.xticks(rotation=90)
plt.show()

# Determine the optimal number of clusters using the dendrogram
# In this case, we visually inspect the dendrogram to identify a suitable number of clusters

# Cut the dendrogram to form clusters
num_clusters = 4  # Adjust as needed
clusters = fcluster(linkage_matrix, num_clusters, criterion='maxclust')

# Add the cluster labels to the DataFrame
df_sorted_confirmed['Cluster'] = clusters

# Print the states and their corresponding clusters
print("States and Their Clusters:")
print(df_sorted_confirmed[['State', 'Cluster']])
```

```python
selected_columns = ['Confirmed', 'Deaths', 'Recovered', 'Active']

# Create a DataFrame with the selected columns
selected_data = df_sorted_confirmed[selected_columns]

# Covariance Matrix
covariance_matrix = selected_data.cov()

# Correlation Matrix (Pearson correlation)
correlation_matrix = selected_data.corr()

# Display the covariance matrix
print("Covariance Matrix:")
print(covariance_matrix)

# Display the correlation matrix
print("\nCorrelation Matrix (Pearson Correlation):")
print(correlation_matrix)
```
